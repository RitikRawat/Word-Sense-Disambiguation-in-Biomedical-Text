```python
i=0
vocab_dict = {}
for key in mwe_final:
    vocab_dict[key] = i
    i=i+1
    if (i == 100000):
        break
```

```python
matrix = np.zeros((len(vocab_dict),len(vocab_dict)))
```

```python
edgelist ={}
for key in vocab_dict:
    edgelist[key] = []
```

```python
ti = 0
for key in MWE:
    dic_lst = MWE[key]
    for dic in dic_lst:
        if len(dic)>1:
            ti+=1
ti
```

```python
for key in MWE:
    dic_lst = MWE[key]
    for dic in dic_lst:
        if len(dic)>1:
            lst = list(dic.keys())
    #         lst2 = []
    #         for token in lst:
    #             if token in vocab_dict:
    #                 lst2.append(token)
    #         if len(lst2)>1:|
            for a, b in itertools.combinations(lst, 2):
                if ((a in vocab_dict) and (b in vocab_dict)):
                    matrix[vocab_dict[a],vocab_dict[b]] += dic[a]*dic[b]
                    matrix[vocab_dict[b],vocab_dict[a]] += dic[a]*dic[b]
```

```python
for row in matrix:
    for item in row:
        if item <3:
            item =0
```
